```python
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
text_data = """
It’s a real pleasure to recommend Antonio. Few professionals demonstrate the level of commitment, integrity, and work ethic that he brings to the table every single day. He’s someone who embraces challenges with enthusiasm and consistently goes the extra mile to deliver results—no matter how difficult the task.

What sets Antonio apart is not only his reliability and sense of responsibility, but also his natural leadership. He earns the respect of everyone around him through transparency, collaboration, and a solutions-oriented mindset. He doesn’t just get things done—he uplifts the whole team while doing it.

Any company would be lucky to have him on board, and any team stronger with him in it.
"""
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))

words = nltk.word_tokenize(text_data.lower())

# remove the stopwords

words = [word for word in words if word.isalnum() and word not in stop_words]

# Create a frequency distribution of the words
worldcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(words))

plt.figure(figsize=(10, 5))
plt.imshow(worldcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```

```python
# Bar Chart of Word Frequencies

word_freq = nltk.FreqDist(words)
most_common_words = word_freq.most_common(20)

plt.figure(figsize=(12, 6))
plt.bar(*zip(*most_common_words), color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram 
plt.hist(word_freq.values(), bins=50, range=[0,50], color='lightgreen', edgecolor='black')
plt.xlabel('Word Frequency')
plt.ylabel('Number of Words')
plt.show()
```
